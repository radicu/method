```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
df = h2o.import_file('data/100_2016/task_data.csv')
```

```python
train, test = df.split_frame(ratios=[0.8])
```

```python
x = df.columns
y = 'TaskDelay'
x.remove(y)
```

```python
aml = H2OAutoML(max_models=20, seed=1, max_runtime_secs=600)
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
print(lb)
```

```python
best_model = aml.leader
```

```python
predictions = best_model.predict(test)
```

```python
h2o.save_model(model=best_model, path='best_model_h2o', force=True)
```
